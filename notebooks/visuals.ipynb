```python
import matplotlib.pyplot as plt
import numpy as np

# Model Performance Comparison Data
models = ["GLMs", "Random Forest", "XGBoost", "Deep Learning (BERT)"]
accuracy = [52.9, 99.6, 93.72, 69.2]
auroc = [0.69, 0.619, 0.70, 0.75]
computational_cost = [1, 3, 4, 5]  # Arbitrary scale (1 = Low, 5 = Very High)

# Creating a bar chart for Accuracy Comparison
plt.figure(figsize=(8, 5))
plt.bar(models, accuracy, color=['blue', 'green', 'red', 'purple'])
plt.xlabel("Models")
plt.ylabel("Accuracy (%)")
plt.title("Model Accuracy Comparison in Insurance Claim Prediction")
plt.ylim(0, 110)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Creating a scatter plot for Computational Cost vs Accuracy Trade-Off
plt.figure(figsize=(8, 5))
plt.scatter(computational_cost, accuracy, color=['blue', 'green', 'red', 'purple'], s=100)
for i, model in enumerate(models):
    plt.text(computational_cost[i] + 0.1, accuracy[i], model, fontsize=10)

plt.xlabel("Computational Cost (Low to High)")
plt.ylabel("Accuracy (%)")
plt.title("Computational Cost vs. Accuracy Trade-Off in ML Models")
plt.xticks(np.arange(1, 6, 1))
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

ML Pipeline for Insurance Claim Prediction

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Create figure with wider aspect ratio
fig, ax = plt.subplots(figsize=(28, 4))

# Pipeline steps - first line (big) and second line (small)
steps = [
    ("DATA COLLECTION", "(Insurance Data)"),
    ("DATA PREPROCESSING", "(Missing Data, Outliers)"),
    ("FEATURE ENGINEERING", "(Encoding & selection)"),
    ("MODEL TRAINING", "(RF, XGBoost, DL)"),
    ("MODEL EVALUATION", "(AUC-ROC, Fairness)"),
    ("FAIRNESS CHECKS", "(Bias Detection, GDPR)"),
    ("PREDICTION", "(Claim Probability)")
]

# Color scheme
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12', '#9b59b6', '#1abc9c', '#34495e']

# Draw boxes and text
for i, ((line1, line2), color) in enumerate(zip(steps, colors)):
    # Draw rectangle
    ax.add_patch(Rectangle((i, 0), 0.9, 0.9, facecolor=color, edgecolor='black', lw=1, alpha=0.8))
    
    # Add text - first line bigger and bold
    ax.text(i + 0.45, 0.6, line1, 
            ha='center', va='center', 
            fontsize=14, fontweight='bold', color='black')
    
    # Add text - second line smaller
    ax.text(i + 0.45, 0.3, line2, 
            ha='center', va='center', 
            fontsize=14, color='black')

# Formatting
ax.set_xlim(-0.5, len(steps) - 0.1)
ax.set_ylim(-0.5, 1.5)
ax.axis('off')


plt.savefig('insurance_ml_pipeline_improved.png', dpi=300, bbox_inches='tight')
plt.show()
```
